# A) Create data subsets

# Subset 1: Posts with more than 100 likes

```python
subset1 = df[df["like"] > 100]
subset1
```

# Subset 2: Posts of type 'Photo'

```python
subset2 = df[df["Type"] == "Photo"]
subset2
```

# Subset 3: Posts published on weekdays only

```python
subset3 = df[df['Post Weekday'] < 6]
subset3
```

# Subset 4: Subset for the columns like, share, and comment

```python
subset4 = df[['Lifetime Post Total Impressions', 'like', 'share', 'comment']]
subset4
```

# Subset 5: Subset for the columns like, Category, and Type

```python
subset5 = df[['Lifetime Post Total Impressions', 'like','Category', 'Type']]
subset5
```

# Subsets Using Slicing (2 subsets)

# Subset 6: First 100 rows

```python
subset6 = df.iloc[:100]
subset6
```

# Subset 7: Select only 'Type' and 'like' columns for all rows

```python
subset7 = df.loc[:, ['Type', 'like']]
subset7
```

# C) Sorting ascending and descending order for single column and multiple columns 


# Sort by Likes in Ascending Order

```python
sorted_data = df.sort_values(by='like', ascending=True)
sorted_data
```

# Sort by Likes in Descending Order

```python
sorted_data = df.sort_values(by='like', ascending=False)
sorted_data
```

```python
sorted_df = df.sort_values(by=['Type', 'like'], ascending=[True, False])
sorted_df
```

# Reshape Data 

# i) Wide to Long Format

```python
melted_data = df.melt(id_vars=['Type'], value_vars=['like', 'share', 'comment'])
melted_data.head()
```

# ii) Long to Wide Format

```python
pivoted_data = melted_data.pivot_table(index='Type', columns='variable', values='value', aggfunc='sum')
pivoted_data.head()
```

# iii) Remove Duplicates Before Pivoting

```python
melted_data = melted_data.drop_duplicates(subset=['Type', 'variable'])
melted_data
```

# iv) Pivot the data

```python
pivoted_data = melted_data.pivot(index='Type', columns='variable', values='value')
pivoted_data
```
