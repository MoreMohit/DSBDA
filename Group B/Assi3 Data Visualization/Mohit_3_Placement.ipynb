# 📊 Data Visualization

```python
# Set style for plots
sns.set_style("whitegrid")

# Create a figure with multiple subplots
# fig, axes = plt.subplots(3, 3, figsize=(18, 15))
```

# 1️⃣ Countplot (Internship vs Placement Status)

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Internship", hue="PlacementStatus", data=df)
plt.title("Internship vs Placement Status")
plt.show()


# This graph shows the number of students who participated in internships and how that impacted their placement.
# If a higher percentage of students with internships are placed, it suggests that internship experience increases placement chances.
# If both categories (with and without internships) have similar placements, internships may not be a strong predictor.
```

# 2️⃣ Scatterplot (CGPA vs Skills (Categorized by Placement Status))

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="CGPA", y="Skills", hue="PlacementStatus", data=df)
plt.title("CGPA vs Skills")
plt.show()

# It visualizes the relationship between CGPA and skills, with colors representing placement status.
# If placed students are concentrated in high CGPA and high skills, it suggests that both are crucial for placement.
# If the distribution is random, placement might depend on other factors.
```

# 3️⃣ Bar Plot (Average CGPA by Placement Status)

```python
    plt.figure(figsize=(8, 5))
    sns.barplot(x="PlacementStatus", y="CGPA", data=df, palette="coolwarm")
    plt.title("Average CGPA by Placement Status")
    plt.show()

#Compares the average CGPA of placed vs. non-placed students.
#If placed students have higher CGPA, it suggests that CGPA is a strong predictor of placement.
```

# 4️⃣ Correlation Heatmap of Numerical Features

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

#  Displays the correlation between numerical features
# Positive values (closer to 1) show strong direct relationships (e.g., CGPA vs 12th Percentage).
# Negative values (closer to -1) indicate inverse relationships.
# A weak correlation (near 0) suggests no significant connection between features
```

# 5️⃣ Histogram

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Communication Skill Rating"], bins=20, kde=True, color="blue")
plt.title("Distribution of Communication Skill Rating")
plt.xlabel("Communication Skill Rating")
plt.ylabel("Frequency")
plt.show()

# Shows the frequency distribution of communication skill ratings
# A right-skewed graph (more students with high ratings) suggests good communication skills overall.
# A normal or left-skewed graph might indicate a varied or generally weaker communication skillset.
```

# 6️⃣ Boxplot

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="PlacementStatus", y="12th Percentage", data=df)
plt.title("12th Percentage vs Placement Status")
plt.show()

# Compares 12th-grade scores of placed vs. non-placed students.
# If placed students have higher 12th Percentage, it suggests that 12th Percentage is a strong predictor of placement.
# If there’s an overlap, 12th Percentage might not be a major factor.
```

# 7️⃣ Pairplot

```python
sns.pairplot(df, vars=["CGPA", "12th Percentage", "Skills"], hue="PlacementStatus", palette="husl")
plt.show()

# Shows relationships between CGPA, 12th Percentage, and Skills, with placement status as a hue.
# Clusters of placed students in the higher CGPA and skill areas suggest these are strong factors.
# If placed students are spread across values, other factors may influence placement.
```

# 8️⃣ Pie Chart

```python
plt.figure(figsize=(6, 6))
df["PlacementStatus"].value_counts().plot.pie(autopct="%1.1f%%", colors=["lightblue", "lightcoral"])
plt.title("Placement Status Distribution")
plt.ylabel("")
plt.show()

# Shows the percentage of students placed vs. not placed.
# A higher proportion of placed students indicates a good placement rate.
# A large proportion of non-placed students suggests possible issues with the selection process or student readiness.
```

# 9️⃣ Violin Plot

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x="PlacementStatus", y="Skills", data=df)
plt.title("Skills vs Placement Status")
plt.show()

# Compares the skill distribution for placed vs. non-placed students
# If the violin is wider for placed students at higher skill values, it suggests skills improve placement chances.
# If distributions are similar, skills may not be a strong determinant
```

# 🔟 Stacked Bar Chart

```python
cross_tab = pd.crosstab(df["Internship"], df["Hackathon"])
cross_tab.plot(kind="bar", stacked=True, figsize=(8, 5), colormap="viridis")
plt.title("Internship vs Hackathon Participation")
plt.xlabel("Internship")
plt.ylabel("Count")
plt.show()

# Shows the number of students who did both or none
# A higher count of students doing both suggests an active engagement in extracurriculars.
# If most students do neither, companies may look for alternative criteria.
```

# 1️⃣1️⃣ LinePlot

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x="12th Percentage", y="CGPA", hue="PlacementStatus", data=df, marker="o")
plt.title("CGPA vs 12th Percentage Trend")
plt.xlabel("12th Percentage")
plt.ylabel("CGPA")
plt.show()

# Shows how CGPA trends with 12th Percentage for different placement statuses.
# If placed students consistently have a higher CGPA and 12th percentage, academic performance is a key factor.
# If the trend varies, placement might depend on other factors like projects and skills.
```
